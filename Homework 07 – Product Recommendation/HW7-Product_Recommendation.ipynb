## Import Library

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## Import Data and Prepare

### Import data

```python
student_df = pd.read_excel("/Users/yo/bads7105/Customer Survey.xlsx")
```

```python
student_df.head()
```

### Delete columns

```python
del student_df["Unnamed: 0"]
del student_df["ประทับเวลา"]
del student_df["Transaction"]
```

### Replace Value (Onehot encoder)

```python
student_df.replace({'เคย':1,'ไม่เคย':0,np.nan:0},inplace = True)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
student_df = student_df.applymap(encode_units)
```

## Algorithm : Apiori for market basket analysis


```python
frequent_itemsets = apriori(student_df, min_support=0.4, use_colnames=True)
```

```python
recommend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
recommend
```

```python
recommend[ (recommend['lift'] >= 1) & (recommend['confidence'] >= 0.5) ]
```
